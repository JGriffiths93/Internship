```python
Brand_name=[]
Name_of_Product=[]
Price=[]
ReturnExchange=[]
Expected_Delivery=[] 
Availability=[]
Product_URL=[]
```

```python
start=0
end=3
for page in range (start,end):
    brand_tags=driver.find_elements(By.XPATH,'//span[@class="a-size-base-plus a-color-base"]')
    for i in brand_tags[0:20]:
        brand=i.text
        Brand_name.append(brand)

    name_tags=driver.find_elements(By.XPATH,'//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]')
    for i in name_tags[0:20]:
        name=i.text
        Name_of_Product.append(name)

    return_tags= driver.find_elements (By.XPATH,"/html/body/div[2]/div/div[5]/div[2]/div[1]/div[2]/div[2]/div/div/div[1]/div[3]/div/div[1]/div[12]/span[2]/span/span/span/a")
    for i in return_tags[0:20]:
        returns=i.text
        ReturnExchange.append(returns)
   
    availability_tags = driver.find_elements (By.XPATH,'//span[@class="a-size-medium a-color-success"]')
    for i in availability_tags[0:20]:
        availability=i.text
        Availability.append(availability)

    delivery_tags= driver.find_elements(By.XPATH,'//span[@class="a-color-base a-text-bold"]')
    for i in delivery_tags[0:20]:
        delivery=i.text
        Expected_Delivery.append(delivery)
    price_tags= driver.find_elements(By.XPATH,'//span[@class="a-price"]')
    for i in price_tags[0:20]:
        price=i.text
        Price.append(price)

    next_button=driver.find_elements(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div[1]/div/span[1]/div[1]/div[66]/div/div/span/a[3]')
    
```
